```python
# does horsepower effect on price
plt.scatter(data['horsepower'],data['price'])
plt.xlabel("horsepower")
plt.ylabel("price")
plt.title("how to horsepower effect on price")
```

```python
# does engine_size effect on price
plt.scatter(data['engine-size'],data['price'])
plt.xlabel("engine-size")
plt.ylabel("price")
plt.title("how to engine-size effect on price")
```

```python
sns.distplot(data['price'])
```

```python
sns.distplot(data['wheel-base'])
```

```python
# Rank cars according to their price
df=data.groupby('make')['price'].agg('mean')
df=pd.DataFrame(df)
df.reset_index(inplace=True)
data_types_car=df.sort_values(by='price',ascending=False)
x_axis=data_types_car['make'][0:10]
y_axis=data_types_car['price'][0:10]
# Visualization settings
sns.set(rc={'figure.figsize':(26,10)}, font_scale=2)
ax = sns.barplot(x_axis, y_axis, palette="Set1")
ax.set(xlabel='Car Brand', ylabel='AVG_Price', title = 'Top 10 cars make based on the price')
for item in ax.get_xticklabels():
    item.set_rotation(45)
```

```python
# compare cars according to their fuel type
df=data.groupby(['fuel-type']).size()#.plot(kind = "bar")
sns.set(rc={'figure.figsize':(8,4)}, font_scale=1.2)
sns.barplot(x = df.index, y = df.values)
```
