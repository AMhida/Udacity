```python
# Import necessary package
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Automobile_data.csv')
```

```python
df = df.replace('?',np.NAN)
```

```python
#Fill quantitave data by mean
cols = ['normalized-losses','bore','stroke','horsepower','peak-rpm','price']
for col in cols:
    df[col] = pd.to_numeric(df[col])
    df[col].fillna(df[col].mean(), inplace = True)
```

```python
bind = int((max(df['price'])-min(df['price']))/3)
bins = range(int(min(df['price'])),int(max(df['price'])),bind)
df['price_bin'] = pd.cut(df['price'],bins,labels=['Low','Med','High'])
```

```python
df.groupby('price_bin')['make'].count()
```

```python
df_group=df.groupby(['price_bin','make']).price.mean().round().dropna()
df_group
```

```python
plt.hist(df['price'])
plt.title('Histegram For Prices')
plt.xticks(rotation = 45 , horizontalalignment="center")
plt.xlabel(' Price ')
plt.ylabel('Number');
```

**Right Skweed most of car in the price interval(5000,15000)**

```python
mean_list=[]
for _ in range(int(1e4)):
    sample_price = np.random.choice(df['price'],205)
    mean_of_sample_price = sample_price.mean()
    mean_list.append(mean_of_sample_price)
plt.hist(mean_list)
plt.title('Histegram For Prices after take CLI')
plt.xticks(rotation = 45 , horizontalalignment="center")
plt.xlabel(' Price ')
plt.ylabel('freq_mean');
```

```python
np.array(mean_list).mean().round()
```
